```python
import os
import re

# Path to the directory containing the images
folder_path = 'path/to/your/folder'

# Regular expression pattern to capture both 'dst' and 'src' prefixes
pattern = re.compile(r'(dst_t|src_t)(\d+\.\d+).png')

# Loop through all files in the directory
for filename in os.listdir(folder_path):
    # Check if the file matches the naming pattern
    match = pattern.match(filename)
    if match:
        # Extract the prefix (dst or src) and the timestamp
        prefix = match.group(1)
        timestamp = float(match.group(2))
        new_timestamp = f"{timestamp:.2f}"  # This will format to two decimal places

        # Construct the new filename
        new_filename = f"{prefix}{new_timestamp}.png"

        # Renaming the file
        os.rename(
            os.path.join(folder_path, filename),
            os.path.join(folder_path, new_filename)
        )
        print(f"Renamed '{filename}' to '{new_filename}'")

# Inform the user the process is complete
print("Renaming process is complete.")
```
